### Tokenizing and Cleaning the reviews 

```python
from nltk.corpus import stopwords
from nltk import word_tokenize
import string

# Tokenizing the cleaned reviews stored in the DataFrame df, 
# converting them to lowercase and removing stopwords and punctuation. 
# The resulting tokens are stored in the list tokens.

stop = set(stopwords.words('english') + list(string.punctuation))
tokens = []
for i in df['Review_cleaned']:
    tokens.append([j for j in word_tokenize(i.lower()) if i not in stop])
print(tokens[0])
```

```python
df['Review_cleaned'] = df['Review_cleaned'].apply(lambda x: [word for word in tokens])
df.head()
```

## WordClouds

```python
Userdf = pd.read_excel("/kaggle/input/userandreviews/User Reviews.xlsx")
Userdf.head()
```

```python
import string
import re

def no_punct(string):
    # Check if the input is a string
    if isinstance(string, str):
        # Remove punctuation
        no_punct_string = re.sub(r'[^\w\s]', '', string)
        return no_punct_string
    else:
        return string  # Return NaN as it is

# Apply the no_punct function to the 'Review' column
Userdf['Review'] = Userdf['Review'].apply(no_punct)
Userdf.head()
```

```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# Convert 'Review' column to string type
Userdf['Review'] = Userdf['Review'].astype(str)

# Filter the DataFrame based on stars
negative_df = Userdf.loc[Userdf['Stars'] <= 3]
positive_df = Userdf.loc[Userdf['Stars'] > 3]

# Concatenate reviews
negative = " ".join(review for review in negative_df['Review'])
positive = " ".join(review for review in positive_df['Review'])
```

```python
stopwords = set(STOPWORDS)
stopwords.update(['read', 'more', 'product'])
```

```python
def generate_wordcloud(data):
    wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(data)

    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
generate_wordcloud(negative)
```

```python
generate_wordcloud(positive)
```
